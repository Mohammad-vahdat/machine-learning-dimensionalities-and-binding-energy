```python
import ase
import ase.io
from ase.io import xsf
import pandas as pd
import json
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np
%matplotlib inline
```

```python
df_labeled_features = pd.read_pickle("features_MultipleFeaturizer.pkl")
df_labeled_features.head()
```

```python
df_binary_labeled_features = df_labeled_features.dropna()
```

```python
df_binary_labeled_features.replace(to_replace='PE', value='EE', inplace=True)
df_binary_labeled_features.replace(to_replace='EE', value=1, inplace=True)
df_binary_labeled_features.replace(to_replace='HBE', value=0, inplace=True)
```

```python
df_binary_labeled_features.head()
```

```python
X = df_binary_labeled_features.drop('label', axis = 1).values
y = df_binary_labeled_features['label'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF = RandomForestClassifier(n_estimators=600)
RF.fit(X_train,y_train)
```

```python
pred_RF = RF.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, pred_RF))
print("\n")
print(confusion_matrix(y_test, pred_RF))
```

```python
with open ('/Users/mohammad/Documents/vdW-radii/Cloned-Repo/machine-learning-dimensionalities-and-binding-energy/data/structures/all_structures_Mounet/export_structures_all_3D_structures_icsd_and_cod_unique.json') as f:
    all_structure = json.load(f)
```

```python
uuids = []
structures = []
i = 0
j = 0
for key in all_structure.keys():
    if key not in labeled_structure.keys():
        try:
            path = 'data/structures/structures_3D_xsf/'

            folder = os.fsencode(path)

            filename = key + '.xsf'

            with open(path + filename) as f:
                structure = ase.io.xsf.read_xsf(f, index=- 1, read_data=False)
        
            uuids.append(key)
            structures.append(structure)
            
            
        except KeyError:
            i += 1
            continue
        except FileNotFoundError:
            j += 1
            continue

print('Number of missing structures due to KeyError:', i)
print('Number of missing structures due to KeyError:', j)
```

```python
data_all_structures = {
    'uuid' : uuids,
    'structure' : structures,
}

dataframe_all_structures = pd.DataFrame(data_all_structures)
```

```python
aa2s = ASEAtomstoStructure()
dataframe_all_structures = aa2s.fit_featurize_dataframe(dataframe_all_structures, col_id= 'structure', ignore_errors=True)
```
